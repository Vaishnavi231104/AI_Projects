```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('stopwords')
#Download required NLTK resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
text = input('Enter a sentence or a paragraph: \n')
```

```python
#1.Tokenization
tokens = word_tokenize(text.lower())
print(tokens)
```

```python
#2. Remove stopwords (common words like 'the', 'is', etc.)
filtered_words=[word for word in tokens if word.isalpha() and word not in stopwords.words('english')]
print(filtered_words)
```

```python
#3. Frequency Distribution
freq=nltk.FreqDist(filtered_words)
print('\n--Word Frequency---')
for word,count in freq.items():
  print(f'{word}:{count}')
```
